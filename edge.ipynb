```
from PIL import Image
import re
import glob
import numpy as np
import os
from sklearn import model_selection
```

```
IMAGE_SIZE = 224
```

```
test_path = './drive/My Drive/Colab Notebooks/AIquest/PBL3_ImageClassification/test/'
train_path = './drive/My Drive/Colab Notebooks/AIquest/PBL3_ImageClassification/train/'
# train = glob.glob(train_path + "**", recursive=True)
# test = glob.glob(test_path + "**", recursive=True)
```

```
classes = ["regular", "potato", "horn", "bridge"]
X = []
Y = []
for index, classlabel in enumerate(classes):
  photos_dir = train_path + classlabel + "/"
  files = glob.glob(photos_dir + "*")
  for i, file_path in enumerate(files):
    file_name = os.path.basename(file_path)
    print(file_name)
    image = Image.open(file_path)
    image.LOAD_TRUNCATED_IMAGES = True
    image = image.resize((IMAGE_SIZE, IMAGE_SIZE))
    data = np.asarray(image)
    X.append(data)
    Y.append(index)

X = np.array(X)
Y = np.array(Y)

X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, Y, test_size=0.2, train_size=0.8
)
xy = (X_train, X_test, y_train, y_test)
np.save(train_path + "/train.npy", xy)
```
